```python
import numpy as np
import matplotlib.pyplot as plt
```

# Line Plot

```python
X = [1,2,3,4,1]
Y = [10,20,30,40,50]
plt.plot(X,Y)
plt.show()
```

# Bar Plot with color

```python
x = np.arange(4)
y = np.random.rand(4)
plt.bar(x,y,color = 'r',width=0.25)
plt.show()
```

```python
x = np.arange(5)
y = np.random.rand(3,5)
plt.bar(x+0.00,y[0],color = 'b',width=0.25)
plt.bar(x+0.25,y[1],color = 'g',width=0.25)
plt.bar(x+0.50,y[2],color = 'r',width=0.25)
plt.show()
```

# Horizontal BAR graph

```python
x = np.arange(4)
y = np.random.rand(4)
plt.barh(x,y)
plt.show()
```

# Scatter plot

```python
x = [1,2,3,4]
y = [1,3,4,2]
plt.scatter(x,y,c='c')
plt.show()
```

```python
x = np.random.rand(20)
y = np.random.rand(20)
colors = np.random.rand(20)
plt.scatter(x, y, c=colors)
plt.show()
```
