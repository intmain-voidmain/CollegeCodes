```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(7)
y1 = x**2
y2 = x**3
y3 = x**4
```

```python
plt.plot(x,y1,label = 'x**2')
plt.plot(x,y2,label = 'x**3')
plt.plot(x,y3,label = 'x**4')
plt.grid(True)
plt.legend(loc='upper center')
plt.show()
```
