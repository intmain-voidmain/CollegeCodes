```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_data = pd.read_csv('district wise rainfall normal.csv')
my_data
```

# Bar Plot annual rainfall for each DISTRICT of SIKKIM, WEST BENGAL, MEGHALAYA

```python
plt.figure(figsize=(10,10))

sikkim = my_data[my_data.STATE_UT_NAME.isin(['SIKKIM'])]
sikkim_x = sikkim.DISTRICT
sikkim_y = sikkim.ANNUAL
plt.barh(sikkim_x,sikkim_y,color = 'r')

wb = my_data[my_data.STATE_UT_NAME.isin(['WEST BENGAL'])]
wb_x = wb.DISTRICT
wb_y = wb.ANNUAL
plt.barh(wb_x,wb_y,color = 'g')

mg = my_data[my_data.STATE_UT_NAME.isin(['MEGHALAYA'])]
mg_x = mg.DISTRICT
mg_y = mg.ANNUAL
plt.barh(mg_x,mg_y,color = 'b')

plt.show()
```

```python
wb1 = my_data[my_data.STATE_UT_NAME.isin(['WEST BENGAL'])]
wb_x = wb1.DISTRICT
wb_y_jun = wb1.JUN
plt.plot(wb_x,wb_y_jun,color = 'g')

wb2 = my_data[my_data.STATE_UT_NAME.isin(['WEST BENGAL'])]
wb_x = wb2.DISTRICT
wb_y = wb2.JUL
plt.plot(wb_x,wb_y,color = 'b')

plt.grid(True)
plt.xticks(rotation = 60)
plt.show()
```

```python
wb3 = my_data[my_data.STATE_UT_NAME.isin(['WEST BENGAL'])]
wb_x = wb3.DISTRICT
jan_feb= wb3['Jan-Feb']
mar_may = wb3['Mar-May']
jun_sep=wb3['Jun-Sep']
oct_dec=wb3['Oct-Dec']
plt.plot(wb_x,jan_feb,color='r')
plt.plot(wb_x,mar_may,color='b')
plt.plot(wb_x,jun_sep,color='g')
plt.plot(wb_x,oct_dec,color='y')
plt.xticks(rotation=60)
plt.grid()
plt.show()

```
