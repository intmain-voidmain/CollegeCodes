# Explore probability distribution and Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
sample_data = np.random.normal(size=100000)
print(sample_data)
plt.hist(sample_data)
plt.show()
sbn.distplot(sample_data)
```

```python
#central limit theorem
def apply_CLT(sample_data,sample_size,total_samples):
    sample_mean = []
    for a in range(total_samples):
        sample = np.random.choice(sample_data,size = sample_size)
        mean = np.mean(sample)
        sample_mean.append(mean)
    return sample_mean
```

```python
sample_data = np.random.normal(size = 100)
sbn.distplot(sample_data)
```

```python
sbn.distplot(apply_CLT(sample_data,30,1000))
```

```python
sample_data = np.random.uniform(size = 10000)
sbn.distplot(sample_data)
```

```python
sbn.distplot(apply_CLT(sample_data,30,100))
```
