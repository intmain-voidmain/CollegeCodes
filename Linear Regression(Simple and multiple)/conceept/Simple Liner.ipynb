```python
import matplotlib.pyplot as plt 
import  pandas as pd
```

```python
df=pd.read_csv("lin1.csv")
df
```

# x_train is independent variable 
#y_train is dependent variable 

```python
x_train=df.drop('Price',axis=1)
y_train=df.drop('Area',axis=1)
print(x_train)
print(y_train)
```

# # Original value represent through  graph 

```python
#x=1 and y=2.1 as per csv file 
import seaborn as sns 
sns.scatterplot(x="Area",y="Price",data=df,color='g',label="original value ")
```

# machine generate our dependent value like x=5 so y = ?  

```python
from sklearn import linear_model 
```

```python
model=linear_model.LinearRegression()
model.fit(x_train,y_train)  # ( independt   , dependt )
```

# y=a+bx
y= dependdent variable 
x=independdent variable
a=intercept ( is the value of dependent when independent variable 
is zero) 
b=regression coefficient or slope (tell the value of change in 
dependent variable with until change independent variable )


```python
#b=coefficent
print(model.coef_)
```

```python
#a=intercept 
print(model.intercept_)
```

# Predict value apply our formula (y=a+bx)

```python
predt=model.predict(x_train)
predt1=x_train
predt1['Price']=predt

sns.scatterplot(x="Area",y="Price",data=predt1,color='r',label="Predict value ")
```

# Analysis Calculation 

y=a+bx
a=2.00244499
b=0.38655257
suppose x=1
y= 2.00244499 + 0.38655257 * 1
y= 2.38899756   # when x is 1 so y is 2.38899756

suppose x=2
y= 2.00244499 + 0.38655257 * 2
y= 2.77555013 # when x is 2 so y is 2.77555013


# Error Find (original value - predict value )

originally , when x is 1 & y = 2.1
predict , when x is 1 & y = 2.38899756

so error is ... 2.1-2.38899756 = -0.2889975599999999

originally , when x is 2 & y = 3.1
predict , when x is 2 & y = 2.77555013

so error is ... 3.1-2.77555013 = 0.32444987000000003


# Graphically represent original value and predict value 

```python
sns.lineplot(x ='Area', y = 'Price', data=predt1,dashes=False,markers = True,color="r" , label="Predict",)

sns.lineplot(x ='Area', y = 'Price', data=df,dashes=False,markers = True,color="g" , label="Original",)
plt.legend()
plt.show()
```

# check How much accurate your model 

```python
x_train=df.drop('Price',axis=1)
y_train=df.drop('Area',axis=1)
model.score(x_train, y_train)
```

```python
# find %of error  
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
predt=model.predict(x_train)
mse = mean_squared_error(y_train, predt) # original value - predict value 
rmse = np.sqrt(mse)

print('MSE = ', mse)
print('RMSE = ', rmse)
```
